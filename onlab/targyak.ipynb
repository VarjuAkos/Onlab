```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://portal.vik.bme.hu/kepzes/targyak/?order=s.code&own=&department_id=vik&has_datasheet=1&active=1&program=all'

page = requests.get(url)

soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
 table = soup.find('table', class_ = 'subject_list')
```

```python
print(table)
```

```python
cleaned_headers = table.find_all('tr')[0]
```

```python
print(cleaned_headers)
```

```python
cleaned_headers.find_all('a')
```

```python
headers = cleaned_headers.text
print(headers)
```

```python
columns = [item for item in headers.split('\n') if item]
print(columns)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = columns)
df
```

```python
data = table.find_all('tr')
row_data = data[2].find_all('td')
print(row_data)
```

```python
for row in data[1:]:
    row_data = row.find_all('td')
    cell = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = cell
```
