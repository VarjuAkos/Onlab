```python
#for few classes:
str = """VIAUAV05
VIAUAV21
VIAUAV23
VIAUAV25
VIAUAV27
VIAUBV18
VIAUAV41
VIAUAV45
VIAUAV49
VIETAV03
VIHIAV30
VIHIAV37
VIHIAV39
VIHIJV47
VIHIAV42
VIHIAV48
VIHIAV51
VIHIAV06
VIHIAV35
VIIIAV16
VIIIAV18
VIMIAV07
VIMIAV12
VIMIAV15
VIMIAV17
VIMIBV09
VIMIJV07
VIMIJV51
VISZA028
VITMAV23
VITMAV42
VITMAV43
VITMBV17
VITMAV82
VIVEAV01"""
classes = str.split('\n')
print(classes)
```

```python
#ez a lenyeg
for individual_class in classes:
    classcode = individual_class
    url_base = 'https://portal.vik.bme.hu/kepzes/targyak/'
    url = url_base + classcode
    page = requests.get(url)
    soup = BeautifulSoup(page.text,'html')

    # CourseName in English
    ps = soup.find_all('p')
    for p in ps:
        text = p.text.strip()
        if 'tantárgy angol neve' in text: 
            #print(text.split(':')[1].strip())
            df.loc[df['Tantárgykód'].str.contains(classcode), 'CourseName'] = text.split(':')[1].strip()

    
    fieldsets = soup.find_all('fieldset')

    for fieldset in fieldsets:
        fieldset_text = fieldset.text.strip()
    
        # Check for specific sections and update the DataFrame accordingly
        if '3. A tantárgyfelelős személy és tanszék' in fieldset_text:
            # Assuming 'classcode' is defined and 'Coordinator' column exists in 'df'
            text = fieldset_text
            name = text.split('\n')[-1].strip(', ')
            df.loc[df['Tantárgykód'].str.contains(classcode), 'Coordinator'] = name
            #print(name)
    
        elif '7. A tantárgy célkitűzése' in fieldset_text:
            # Update another column, assuming its name is 'Objectives'   
            obj = fieldset_text.replace("7. A tantárgy célkitűzése", "").strip()
            #print(obj)
            df.loc[df['Tantárgykód'].str.contains(classcode), 'Objectives'] = obj  
    
        elif '8. A tantárgy részletes tematikája' in fieldset_text:
            # Update another column, assuming its name is 'DetailedTopics'
            dt = fieldset_text.replace("8. A tantárgy részletes tematikája", "").strip()
            cleaned_text = re.sub(r'\s+', ' ', dt)
            df.loc[df['Tantárgykód'].str.contains(classcode), 'DetailedTopics'] = cleaned_text
            #print(cleaned_text)
        
        elif '9. A tantárgy oktatásának módja' in fieldset_text:
            # Update another column, assuming its name is 'DetailedTopics'
            dt = fieldset_text.replace("9. A tantárgy oktatásának módja (előadás, gyakorlat, laboratórium)", "").strip()
            cleaned_text = re.sub(r'\s+', ' ', dt)
            df.loc[df['Tantárgykód'].str.contains(classcode), 'MethodOfInstruction'] = cleaned_text
            #print(cleaned_text)

        elif '10. Követelmények' in fieldset_text:
            # Update another column, assuming its name is 'DetailedTopics'
            dt = fieldset_text.replace("10. Követelmények", "").strip()
            cleaned_text = re.sub(r'\s+', ' ', dt)
            df.loc[df['Tantárgykód'].str.contains(classcode), 'Assessment'] = cleaned_text
            #print(cleaned_text)

        elif '11. Pótlási lehetőségek' in fieldset_text:
            # Update another column, assuming its name is 'DetailedTopics'
            dt = fieldset_text.replace("11. Pótlási lehetőségek", "").strip()
            cleaned_text = re.sub(r'\s+', ' ', dt)
            df.loc[df['Tantárgykód'].str.contains(classcode), 'Recaps'] = cleaned_text
            #print(cleaned_text)

        elif '12. Konzultációs lehetőségek' in fieldset_text:
            # Update another column, assuming its name is 'DetailedTopics'
            dt = fieldset_text.replace("12. Konzultációs lehetőségek", "").strip()
            cleaned_text = re.sub(r'\s+', ' ', dt)
            df.loc[df['Tantárgykód'].str.contains(classcode), 'Consultations'] = cleaned_text
            #print(cleaned_text)
        elif '13. Jegyzet, tankönyv' in fieldset_text:
            # Update another column, assuming its name is 'DetailedTopics'
            dt = fieldset_text.replace("13. Jegyzet, tankönyv, felhasználható irodalom", "").strip()
            cleaned_text = re.sub(r'\s+', ' ', dt)
            df.loc[df['Tantárgykód'].str.contains(classcode), 'Resources'] = cleaned_text
            #print(cleaned_text)
        elif '14. A tantárgy elvégzéséhez' in fieldset_text:
            # Update another column, assuming its name is 'DetailedTopics'
            dt = fieldset_text.replace("14. A tantárgy elvégzéséhez átlagosan szükséges tanulmányi munka", "").strip()
            cleaned_text = re.sub(r'\s+', ' ', dt)
            df.loc[df['Tantárgykód'].str.contains(classcode), 'RequiredLearningHours'] = cleaned_text
            #print(cleaned_text)


#df2 = df[df['Coordinator'].notna() & (df['Coordinator'] != 'nan')]
df
```
