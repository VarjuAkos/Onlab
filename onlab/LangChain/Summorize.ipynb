```python
def get_completion(prompt):
    response = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        model="gpt-3.5-turbo"
    )
    return response.choices[0].message.content
print(get_completion("what is 1+1"))
```

## LLMChain


```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.chains import LLMChain
from langchain.chains import SequentialChain
from langchain_openai import OpenAIEmbeddings
from langchain_community.llms import openai
```

```python

#llm = ChatOpenAI(temperature=0.9, model = "gpt-4-0125-preview")
llm = ChatOpenAI(temperature=0.9)
```

```python
# Translate
translate_prompt = ChatPromptTemplate.from_template(
    "Translate the following to english:"
    "{Text}"
)

translate_chain = LLMChain(llm = llm, prompt = translate_prompt)
for index, row in df.iterrows():
    
    objectives = translate_chain.run(row['Objectives'])
    df.at[index, 'Objectives-en'] = objectives
    #print(objectives)
    
    detailedTopics = translate_chain.run(row['DetailedTopics'])
    df.at[index, 'DetailedTopics-en'] = detailedTopics
    #print(detailedTopics)
    
    methodOfInstruction = translate_chain.run(row['MethodOfInstruction'])
    df.at[index, 'MethodOfInstruction-en'] = methodOfInstruction
    #print(methodOfInstruction)
    
    assessment = translate_chain.run(row['Assessment'])
    df.at[index, 'Assessment-en'] = assessment
    #print(assessment)
    
    recaps = translate_chain.run(row['Recaps'])
    df.at[index, 'Assessment-en'] = recaps
    #print(recaps)
    
    consultations = translate_chain.run(row['Consultations'])
    df.at[index, 'Consultations-en'] = consultations
    #print(consultations)
    
    requiredLearningHours = translate_chain.run(row['RequiredLearningHours'])
    df.at[index, 'RequiredLearningHours-en'] = requiredLearningHours
    #print(requiredLearningHours)
  
```

```python
# prompt template 1: translate to english
first_prompt = ChatPromptTemplate.from_template(
    "Translate the following Objectives and detail topics to english:"
    "\n\n"
    "Objectives:{Obj_hu}"
    "Detailed topics:{Dt_hu}"
)
# chain 1: input= Review and output= English_Review
chain_one = LLMChain(llm=llm, prompt=first_prompt, output_key="Translation")
```

```python
# prompt 2: summarize the following 
second_prompt = ChatPromptTemplate.from_template(
    "Your task as a University professor to make a short review and evaluation of the available courses, the Objectives, and the detailed topic is too long for the lazy students."
    "Make a summarization of the following and reduce size so it is more understandable for future students"
    "\n\n"
    "Objectives and detailed topics:{Translation}"
)
# chain 2: input= English_Review and output= summary
chain_two = LLMChain(llm=llm, prompt=second_prompt, output_key="summary")
```

```python
# prompt template 3: Create a review
third_prompt = ChatPromptTemplate.from_template(
    "Given the following course information:"
    "Course Code: {CourseCode}"
    "Course Name: {CourseName}"
    "Coordinator: {Coordinator}"
    "Course Site Path: {CourseSitePath}"
    "Department:{Department}"
    "Summary: {summary}"
   
    "Generate a concise review that includes:"
    "1. A brief overview of what the course is about."
    "2. Key learning outcomes and skills students can acquire."
    "3. The advantages of taking this course."
    "4. The relevance of the course content to current market trends and technology advancements."
    "5. Exciting projects or technologies students can explore with the skills learned in this course."
)
# chain 3: input= Review and output= language
chain_three = LLMChain(llm=llm, prompt=third_prompt,output_key="ShortReview")
```

```python
overall_chain = SequentialChain(
    chains=[chain_one, chain_two, chain_three],
    input_variables=["Obj_hu","Dt_hu","CourseCode","CourseName","Coordinator","CourseSitePath","Department"],
    output_variables=["Translation", "summary","ShortReview"],
    verbose=False
)
```
